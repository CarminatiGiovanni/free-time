```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import scipy as sc
```

```python
# classical potential
A,B = 1,1
r0 = 2*B/A
L = 10

V = lambda r: -A/r + B/(r**2)
Rin = np.array([r0,1.5,0.8])
E = V(Rin)

x = np.linspace(0.05,4,1000)
plt.plot(x,V(x))
plt.axhline(0, color='gray', linestyle='--')
plt.axhline(E[0], color='lime', linestyle='--',label='E='+str(np.round(E[0],2)))
plt.axhline(E[1], color='red', linestyle='--',label='E='+str(np.round(E[1],2)))
plt.axhline(E[2], color='yellow', linestyle='--',label='E='+str(np.round(E[2],2)))
plt.title('Classical potential')
plt.xlabel('r')
plt.ylabel('V(r)')
plt.ylim(-0.5,1)
plt.legend()
plt.show()
```
